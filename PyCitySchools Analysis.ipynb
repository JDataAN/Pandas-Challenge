```python

# Create a DataFrame called `per_school_summary` with columns for the calculations above.
school_data = {
    "School Names": [
        "Bailey High School", "Cabrera High School", "Figueroa High School",
        "Ford High School", "Griffin High School", "Hernandez High School",
        "Holden High School", "Huang High School", "Johnson High School",
        "Pena High School", "Rodriguez High School", "Shelton High School",
        "Thomas High School", "Wilson High School", "Wright High School"
    ],
    "Students Passing Math": [
        3318, 1749, 1946, 1871, 1371, 3094, 395, 1916, 3145, 910, 2654, 1653, 1525, 2143, 1680
    ],
    "Students Passing Reading": [
        4077, 1803, 2381, 2172, 1426, 3748, 411, 2372, 4867, 923, 3208, 1688, 1591, 2204, 1739
    ],
    "Students Passing Math and Reading": [
        2719, 1697, 1569, 1487, 1330, 2481, 381, 1561, 2549, 871, 2119, 1583, 1487, 2068, 1626
    ]
}

# Create a DataFrame from the school data
school_df = pd.DataFrame(school_data)

# Numeric Data
school_df[['Students Passing Math', 'Students Passing Reading']] = school_df[['Students Passing Math', 'Students Passing Reading']].astype(int) 

# Total number of students per school 
school_df['Total Students'] = school_df[['Students Passing Math', 'Students Passing Reading']].sum(axis=1)  

# Per School Passing Rates (Math and Reading) 
school_df['Per School Math Pass Rate'] = school_df['Students Passing Math'] / school_df['Total Students'] * 100  
school_df['Per School Reading Pass Rate'] = school_df['Students Passing Reading'] / school_df['Total Students'] * 100

# Overall Passing Rate 
school_df['Overall Passing Rate'] = school_df['Students Passing Math and Reading'] / school_df['Total Students'] * 100

# per_school_summary DataFrame 
per_school_summary = school_df[['School Names', 'Total Students', 'Per School Math Pass Rate', 'Per School Reading Pass Rate', 'Overall Passing Rate']]

# Formatting
rounded_rates = per_school_summary[['Per School Math Pass Rate', 'Per School Reading Pass Rate', 'Overall Passing Rate']].round(1)  # Match column names exactly
per_school_summary[['Per School Math Pass Rate', 'Per School Reading Pass Rate', 'Overall Passing Rate']] = rounded_rates


# Display
per_school_summary
```

```python
# Sort the schools by `% Overall Passing` in descending order and display the top 5 rows.

#Merge school_data_complete with school_df 
merged_data = pd.merge(school_data_complete, school_df, how='left', left_on='school_name', right_on='School Names')

# Calculate
merged_data['% Passing Math'] = (merged_data['math_score'] >= 70) * 100
merged_data['% Passing Reading'] = (merged_data['reading_score'] >= 70) * 100
merged_data['% Overall Passing'] = ((merged_data['math_score'] >= 70) & (merged_data['reading_score'] >= 70)) * 100

# Group by school_name 
grouped_data = merged_data.groupby('school_name')

# DataFrame 
summary_df = pd.DataFrame({
    'School Type': grouped_data['type'].first(),
    'Total Students': grouped_data['Student ID'].count(),
    'Total School Budget': grouped_data['budget'].first().map('${:,.2f}'.format),
    'Per Student Budget': grouped_data['budget'].first() / grouped_data['Student ID'].count(),
    'Average Math Score': grouped_data['math_score'].mean(),
    'Average Reading Score': grouped_data['reading_score'].mean(),
    '% Passing Math': grouped_data['% Passing Math'].mean(),
    '% Passing Reading': grouped_data['% Passing Reading'].mean(),
    '% Overall Passing': grouped_data['% Overall Passing'].mean()
}).reset_index()

# Sort the DataFrame by '% Overall Passing' in descending order
summary_df = summary_df.sort_values(by='% Overall Passing', ascending=False)

# Display the top 5 schools by overall passing rate
top_schools = summary_df.head(5)
top_schools
```
